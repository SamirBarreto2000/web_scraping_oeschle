```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.oechsle.pe/tecnologia/computo/laptops-gamers'
r = requests.get(url)

r.status_code
```

```python
soup = BeautifulSoup(r.content, 'html.parser')
```

```python
nombre_equipo = soup.find_all('span', attrs={"class":"text fz-15 fz-lg-17 prod-name"})
nombre_equipo = [i.text for i in nombre_equipo]



precio_actual = soup.find_all('span', attrs={"class": "text fz-lg-15 fw-bold BestPrice"})
precio_actual = [i.text.replace('S/. ', '') for i in precio_actual]



marca_equipo = soup.find_all('p', attrs={"class": "texto"})
marca_equipo = [i.text.upper() for i in marca_equipo]


precio_anterior = soup.find_all('div', attrs={"class": "col-9 text-right"})
precio_anterior = [i.text.replace('\nS/. ', '').replace('\n', '') for i in precio_anterior]


precio_tarjeta_oh = soup.find_all('span', attrs={"class": "text text-secondary fz-lg-15 fw-bold tOhPrice"})
precio_tarjeta_oh = [i.text for i in precio_tarjeta_oh]

```

```python
import pandas as pd 
```

```python
 
data_equipos = {
    'Nombre Equipo': nombre_equipo, 
    'Precio Actual': precio_actual,
    'Precio Anterior': precio_anterior,
    'Marca': marca_equipo
} 
df_equipos = pd.DataFrame(data_equipos) 
df_equipos
```

```python
ruta_archivo_excel = 'laptops-gamers_ripley.xlsx' 
df_equipos.to_excel(ruta_archivo_excel, index=False)  
print(f"Archivo Excel '{ruta_archivo_excel}' creado exitosamente.")
```
